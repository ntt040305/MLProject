```python
# Import thư viện
import pandas as pd
from decision_tree_c45.src.preprocessing import preprocess, detect_types
from decision_tree_c45.src.decision_tree import DecisionTreeC45
from decision_tree_c45.src.visualization import print_tree
```

```python
# Nạp dữ liệu
df = pd.read_csv('data/car_price_prediction_.csv')
df.head()
```

```python
# Chọn mục tiêu và tiền xử lý
target = 'Condition'
df_clean, features, types = preprocess(df, target, drop_cols=['Car ID', 'Model'])
features, types
```

```python
# Huấn luyện cây
X = df_clean[features]
y = df_clean[target]
tree = DecisionTreeC45(feature_types=types, max_depth=8)
tree.fit(X, y)
print_tree(tree.root)
```
